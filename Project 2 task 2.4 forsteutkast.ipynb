$$

    (\mu, w) - (\gamma (\phi^3 - \phi), w) - (\epsilon \nabla \phi, \nabla w) &= (g, w) \quad \text{for all } w \in V.
\end{align}

To simplify the study we choose to implement Dirchlet boundry conditions. To test the problem we choose the solutions;

The expressions for $\phi^3 - \phi$ and \(\Delta \phi\) are computed as follows:
$$

$$

The Laplacian of $\phi$ is:
$$
\Delta \phi = \frac{\partial^2 \phi}{\partial x^2} + \frac{\partial^2 \phi}{\partial y^2}.
$$
First we compute the second derivatives:
$$
\frac{\partial^2 \phi}{\partial x^2} = 2y(y-1)2t, \quad \frac{\partial^2 \phi}{\partial y^2} = 2x(x-1)2t.
$$
Thus:
$$
\Delta \phi = 4t \big(y(y-1) + x(x-1)\big).
$$

We are also asked check how the rate of disspersion compares for the different schemes. We can calculate this from the deffinition of the energy; 
$$
E(\phi(t)) = \int_{\Omega} \frac{\epsilon}{2} | \nabla \phi|^2 +\frac{\gamma}{4}(\phi^2-1)^2dx.
$$

The disspersion rate is deffined as the derivative, which we found an expression for in task $2.2.3$;

$$
\frac{\partial E}{\partial t} = - \int_{\Omega} M |\nabla \mu|^2  d\Omega
$$

For our test problem we have defined 

$$
\mu(x, y, t) = \cos(\pi x) \cos(\pi y) e^{-t}.
$$

We calculate the parital derivative of $\mu$; 

$$
\nabla \mu = \left[-\pi \sin(\pi x) \cos(\pi y)e^{-t}, -\pi\cos(\pi x) \sin(\pi y)e^{-t}\right]
$$

$$
\implies \nabla \mu \cdot \nabla \mu  = \pi^2e^{-2t}\left[ (\sin(\pi x) \cos(\pi y))^2 + (\cos(\pi x) \sin(\pi y))^2\right]
$$

Here we can use a trigonometric identity; 
$$
(\sin A \cos B)^2 + (\cos A \sin B)^2 = (\sin^2 A)(\cos^2 B) + (\cos^2 A)(\sin^2 B).
$$

Using this identity:
$$
(\sin A \cos B)^2 + (\cos A \sin B)^2 = \sin^2(A)(1 - \sin^2(B)) + (1 - \sin^2(A))\sin^2(B).
= \sin^2(A) + \sin^2(B).
$$

This gives the  expression;

$$
|\nabla \mu|^2 = \pi^2e^{-2t}\left[ \sin^2(\pi x) + \sin^2(\pi y)\right]
$$

We substitute this into the integral exprssion; 

$$
\begin{align*}
\frac{\partial E}{\partial t} &= - \int_{\Omega} M \pi^2e^{-2t}\left[ \sin^2(\pi x) + \sin^2(\pi y)\right]  d\Omega \\
& = -\int_{0}^{1}\int_{0}^{1} M \pi^2e^{-2t}\left[ \sin^2(\pi x) + \sin^2(\pi y)\right]  dxdy \\ 
& = -\pi^2 M e^{-2t}
\end{align*}
