### Text theory for task 2.4

We are asked to check the correctness of $\phi$, $\mu$ for our solvers. We will do this by manufactoring solutions, and their respective right-hand sides $f,g$, such that 

$$
\begin{align}
    (\partial_t \phi, v) + (M \nabla \mu, \nabla v) &= (f, v) \quad \text{for all } v \in V, \\
    (\mu, w) - (\gamma (\phi^3 - \phi), w) - (\epsilon \nabla \phi, \nabla w) &= (g, w) \quad \text{for all } w \in V.
\end{align}
$$

To simplify the study we choose to implement Dirchlet boundry conditions. To test the problem we choose the solutions;
$$
\phi(x, y, t) = x(x-1)y(y-1)2t.
$$

The expressions for $\phi^3 - \phi$ and \(\Delta \phi\) are computed as follows:
$$
\phi^3 - \phi = \big(x(x-1)y(y-1)2t\big)^3 - x(x-1)y(y-1)2t,
$$
$$
\phi^3 - \phi = 8t^3 \big(x(x-1)y(y-1)\big)^3 - 2t \big(x(x-1)y(y-1)\big).
$$

The Laplacian of $\phi$ is:
$$
\Delta \phi = \frac{\partial^2 \phi}{\partial x^2} + \frac{\partial^2 \phi}{\partial y^2}.
$$
First we compute the second derivatives:
$$
\frac{\partial^2 \phi}{\partial x^2} = 2y(y-1)2t, \quad \frac{\partial^2 \phi}{\partial y^2} = 2x(x-1)2t.
$$
Thus:
$$
\Delta \phi = 4t \big(y(y-1) + x(x-1)\big).
$$

The expression for $\mu$ is:
$$
\mu = \gamma (\phi^3 - \phi) - \epsilon \Delta \phi,
$$
$$
\mu = \gamma \Big[8t^3 \big(x(x-1)y(y-1)\big)^3 - 2t \big(x(x-1)y(y-1)\big)\Big] - 4\epsilon t \big(y(y-1) + x(x-1)\big).
$$

The right-hand side $f$ is computed using:
$$
f = \frac{\partial \phi}{\partial t} + \nabla \cdot (M \nabla \mu).
$$
The time derivative of $\phi$ is:
$$
\frac{\partial \phi}{\partial t} = x(x-1)y(y-1)2.
$$

$$
