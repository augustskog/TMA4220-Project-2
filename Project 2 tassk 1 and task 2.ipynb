### Task 2.1 

We are given the fourth order parabolic equation; 
\begin{equation}
\partial_t \phi = \nabla \cdot \left[ M \nabla \left( \gamma (\phi^3 - \phi) - \epsilon \Delta \phi \right) \right] \quad \text{in } \Omega \times [0, T],
\end{equation}

where $M, \epsilon > 0$. Becuase of the high complexity, we choose to split the introduce a new variable to split the problem into two coupled second-order problems. We are asked to show that the problem could be rewritten as 

\begin{align*}
\partial_t \phi = \nabla \cdot \left[ M \nabla \mu \right] \quad \text{in } \Omega \times [0, T] \\
\mu = \gamma (\phi^3 - \phi) - \epsilon \Delta \phi \quad \text{in } \Omega \times [0, T],
\end{align*}

where $\mu$ can be written as the chemical potential driving the evolution of the equation. To show this we start by defining the expression for $\mu$; 


\begin{align*}
\mu = \gamma (\phi^3 - \phi) - \epsilon \Delta \phi \quad \text{in } \Omega \times [0, T].
\end{align*}

This is a nonlinear, second order partial differential equation on the space $\Omega$. By simply substituting the definition of $\mu$ back into the original equation, we get the parabolic equation;
\begin{align*}
\partial_t \phi = \nabla \cdot \left[ M \nabla \mu \right] \quad \text{in } \Omega \times [0, T]. \\
\end{align*}

### 1 
We are now asked to derive the weak formulation based on the expressions found in task $2.1$. The weak formulation can bw written as: find $(\phi, \mu) \in \mathbb{V}$ such that 
\begin{align*}
(\partial_t \phi, v) + (M \nabla \mu, \nabla v) &= 0 \quad \text{for all } v \in V \\
(\mu, w) - (\gamma (\phi^3 - \phi), w) - (\epsilon \nabla \phi, \nabla w) &= 0 \quad \text{for all } w \in V
\end{align*}



We start by showing that this holds true for the time dependent expression from task $2.1$. We multiplying the expression by a function $v \in \mathbb{V}$, and integrate the product over $\Omega$;
$$
\int_{\Omega} \phi_t v d \Omega -\int_{\Omega}  \nabla \cdot \left[ M \nabla \mu \right] v d \Omega = \int_{\Omega} \phi_t v d \Omega -\int_{\Omega} M \Delta \mu v d \Omega = 0
$$

We use the Greens forumla for the Laplacian, which states the following; 

$$
-\int_{\Omega} \Delta u v d \Omega = \int_{\Omega} \nabla u \nabla v d\Omega - \int_{\partial \Omega} \frac{\partial u}{\partial n} v d\gamma,
$$
where $\partial \Omega$ is the boundry of the domain $\Omega$, and $n$ is the outward normal unit vector on $\partial \Omega$. We swap this relation into the expression, and get the following 

$$
\int_{\Omega} \phi_t v d \Omega -\int_{\Omega}  \nabla \cdot \left[ M \nabla \mu \right] v d \Omega  = \int_{\Omega} \phi_t v d \Omega + M \left[\int_{\Omega} \nabla \mu \nabla v d\Omega - \int_{\partial \Omega} \frac{\partial \mu}{\partial n} v d\gamma \right] = 0
$$

For the expression to match the target expression, we need the boundry term to be equal to zero. This holds true if we assume Neumann boundry conditions, i.e. $\frac{\partial \mu}{\partial n}|_{\partial \Omega} = 0$. To ensure that the integral does not converge, we need $v \in H^1$ aswell. This gives the following result; 

$$
\int_{\Omega} \phi_t v d \Omega -\int_{\Omega}  \nabla \cdot \left[ M \nabla \mu \right] v d \Omega  = \int_{\Omega} \phi_t v d \Omega + M \int_{\Omega} \nabla \mu \nabla v d\Omega = (\partial_t \phi, v) + (M \nabla \mu, \nabla v) = 0,
$$

where $\mu \in \mathbb{V} = \left\{ \mu \in H^1(\Omega) : \nabla \mu \cdot n = 0 \right\}$, where $n$ is the outward normal unit vector on $\partial \Omega$. This can be thought of as a mass conservation condition, since the derivative out of the area is always zero.

Now we need to prove the same for the nonelinear second order equation. We multiplying the expression by a function $w \in \mathbb{V}$, and integrate the product over $\Omega$;
$$
\int_{\Omega}\mu w d\Omega- \int_{\Omega}\gamma (\phi^3 + \phi)w d\Omega -\int_{\Omega} \epsilon \Delta \phi w d\Omega=0
$$
We again use Greens formula for the Laplacian on the last term in the expression, and get the following; 
$$
\int_{\Omega}\mu w d\Omega- \int_{\Omega}\gamma (\phi^3 - \phi) w d\Omega - \int_{\Omega} \epsilon \nabla \phi \nabla w d\Omega + \int_{\partial \Omega} \epsilon \nabla \phi \cdot n w d\gamma =0
$$

We observe that since $\phi \in \mathbb{V}$, the last term will be equal to zero. We therefore get the following expression; 

$$
\int_{\Omega}\mu w d\Omega- \int_{\Omega}\gamma (\phi^3 - \phi) w d\Omega - \int_{\Omega} \epsilon \nabla \phi \nabla w d\Omega =(\mu, w) - (\gamma (\phi^3 - \phi), w) - (\epsilon \nabla \phi, \nabla w) = 0,
$$

where we agian note that $\phi \in \mathbb{V}$

We are asked to prove that the total mass - $\int_{\Omega} \phi(x,t) dx = C$ - is conserved in time, under the conditions stated in 1. To prove this we simply take the time derivative of the mass expression, and use the original equations. 

$$
M(t) = \int_{\Omega} \phi(x,t) dx
$$

$$
\frac{\partial M(t)}{\partial t} = \int_{\Omega} \frac{\partial \phi(x,t)}{\partial t} dx = \int_{\Omega} \nabla \cdot \left[ M \nabla \mu \right] dx 
$$

We use the divergenve theorem on Lagrangian, which states; 

$$
\int_{\Omega} \nabla \vec{F} d\Omega = \int_{\partial \Omega} \vec{F} \cdot n d\gamma
$$

We substitute this in for the last expression, and get; 
$$
\frac{\partial M(t)}{\partial t} =  \int_{\partial \Omega} M \nabla \mu \cdot n d\gamma = \int_{\partial \Omega} M \frac{\partial \mu}{\partial n} d\gamma = 0.
$$

We claim that this is equal to zero since we know that $\mu \in \mathbb{V}$, which means that $\frac{\partial \mu}{\partial n} = 0$ on $\partial \Omega$. Since we know the derivative of the function is zero, this means that the original function must be a constant.

$$
\frac{\partial M(t)}{\partial t} = 0, \implies M(t) = c,
$$


Physically, this tells us that the total mass in the system is preserved over time. This makes sense considering we assume no boundry flux with the neumann conditions.

We are given the energy norm 
$$
E(\phi(t)) = \int_{\Omega} \frac{\epsilon}{2} | \nabla \phi|^2 +\frac{\gamma}{4}(\phi^2-1)^2dx,
$$

and asked to show that it decreases with time. We do this by showing that the time derivative is always negative. 

$$
\begin{align*}
\frac{\partial E}{\partial t} &= \int_{\Omega} \frac{\epsilon}{2} \cdot 2\cdot |\nabla \phi | \cdot \phi_t + \frac{\gamma}{4}\cdot 2\cdot (\phi^2-1) \cdot 2\phi \cdot \phi_t dx\\
&=\int_{\Omega} \epsilon \nabla \phi \nabla \phi_t + \gamma (\phi^3-\phi)\phi_tdx
\end{align*}
$$

Now we use Greens relation and the assumption of no flux, i.e. $\frac{\partial \mu}{\partial n} = 0$ on $\partial \Omega$.

$$
\int_{\Omega} \nabla \phi \nabla \phi_t dx = - \int_{\Omega} \Delta \phi \phi_t dx +\int_{\partial \Omega} \frac{\partial \mu}{\partial n} \phi_t d \gamma = - \int_{\Omega} \Delta \phi \phi_t dx
$$

$$
\begin{align*}
\implies \frac{\partial E}{\partial t} &= \int_{\Omega} - \epsilon \Delta \phi \phi_t +  \gamma (\phi^3-\phi)\phi_t dx\\
&= \int_{\Omega} \left[- \epsilon \Delta \phi +  \gamma (\phi^3-\phi)\right]\phi_t dx
= \int_{\Omega} \mu \phi_t dx
\end{align*}
$$

We can now use the time dependent equation we found in task 2.1 to swap $ \phi_t $ with $ \nabla \cdot \left[ M \nabla \mu  \right]$, and we will use the green relation on the expression again under the no boundry flux assumption.  

$$
\begin{align*}
\frac{\partial E}{\partial t} &= \int_{\Omega} \mu \phi_t dx = \int_{\Omega} \mu \nabla \cdot \left[ M \nabla \mu  \right] dx = \int_{\Omega} \mu M \Delta \mu  dx \\
&= - \int_{\Omega} M \nabla \mu  \nabla \mu  dx = - \int_{\Omega} M |\nabla \mu|^2  dx \leq 0
\end{align*}
$$


Since the derivative of the energy function is always negative, we can conclude that the energy must be decreasing in time.
